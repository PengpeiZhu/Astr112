```python
import matplotlib.pyplot as plt
import pandas
import numpy as np
%matplotlib inline
```

```python
data = pandas.read_csv('HW1-4-data.csv')
```

```python
data
```

```python
cut1 = (data['parallax']<2)
cut2 = (data['parallax']>0)
```

```python
%matplotlib notebook
plt.figure(figsize=(10,7))
plt.hist(data['parallax'][cut1 & cut2],bins = 500)
plt.xlabel('parallax[kpc]')
plt.ylabel('number of stars')
```

```python
cut3 = (data['parallax']<1.20)
cut4 = (data['parallax']>1.09)

plt.figure(figsize=(10,7))

plt.hist(data['parallax'][cut1 & cut2],bins = 700)
plt.hist(data['parallax'][cut3 & cut4],bins = 40, label = 'M67, 1.09-1.20[kpc]')
plt.xlabel('parallax[kpc]')
plt.ylabel('density of stars')
plt.title('HW1_problem4_(a)')
plt.legend()
plt.show()
plt.savefig('HW1_problem4_(a).png')
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data['bp_rp'][cut3 & cut4],data['phot_g_mean_mag'][cut3 & cut4],s=1)
plt.ylim(22,7)
plt.xlabel('(BP-RP)_0')
plt.ylabel('Magnitude in G')
plt.title('HW1_problem4_(b)')
plt.savefig('HW1_problem4_(b).png')
```
